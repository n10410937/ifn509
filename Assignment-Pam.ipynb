```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

#Chart
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#https://github.com/n10410937/ifn509/raw/master/time_series_19-covid-Confirmed.csv
path='https://github.com/n10410937/ifn509/raw/master/time_series_19-covid-'
status=['Confirmed', 'Deaths', 'Recovered']
dfs=[] # list contains serveral dataframes
for s in status:
    d= pd.read_csv(f'{path}{s}.csv')
#     d.to_csv(f'covid19_{s}.csv', index=False)
    dfs.append(d)
```

```python
#append data
df=pd.concat(dfs, keys=status)
df
```

```python
#unpivot data
dtmps=[]
for k in status:
    dtmps.append(pd.melt(df.loc[k], 
            id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
            var_name='Date', value_name=k))
```

```python
dcovid=pd.concat(dtmps, axis=1)
dcovid.columns.duplicated() 
dcovid=dcovid.loc[:, ~dcovid.columns.duplicated()] # dalate duplicated column
dcovid
```

```python
#covert date data type
dcovid['Date']=pd.to_datetime(dcovid['Date'])
```

```python
#sort data
dcovid=dcovid.sort_values(['Country/Region', 'Province/State', 'Date']).reset_index(drop=True)
dcovid
```

```python
#group by country
dcountry=dcovid.groupby(['Country/Region', 'Date']).sum()[['Confirmed', 'Deaths', 'Recovered']]
dcountry.reset_index(inplace=True)
dcountry
```

```python
dcountry['daily_new_patient']=dcountry.groupby(['Country/Region'])['Confirmed'].diff()
dcountry['daily_new_patient_pct']=dcountry.groupby(['Country/Region'])['Confirmed'].pct_change()
dcountry['Active'] = dcountry['Confirmed']- dcountry['Deaths']- dcountry['Recovered']
dcountry
```

## How many countries have reported at least 10 cases?

```python
dq=dcountry[['Country/Region', 'Confirmed']].groupby('Country/Region').max() > 10
print("Countries have reported at least 10 cases: ")
dq.count()
```

## Which are the five countries with the highest number of active cases?

```python
dq=dcountry[['Country/Region', 'Active']].groupby('Country/Region').max()
dq.sort_values(by=['Active'],ascending=False,inplace=True)
dq[:5]
```

## What is the current rate of increase in the total number of cases, based on the last week of data?

```python
i = len(df.loc['Confirmed'].columns)
j = i-7
sum_col = df.loc['Confirmed'].sum(axis=0)
rate = (sum_col[i-3]-sum_col[j-3])/sum_col[j-3]
print("The current rate of increase in the total number of cases is: ", rate)
```

```python
dq=dcountry[['Country/Region', 'Confirmed']].groupby('Country/Region').max() > 300
dq.index[dq.Confirmed]

@interact(country=pd.Series(dq.index[dq.Confirmed]))
def plot_chart(country):
    df=dcountry[dcountry['Country/Region'].str.contains(country)]
    df[['Date', 'Confirmed', 'Recovered', 'Deaths']].set_index('Date').plot(figsize=(16, 7), 
                                                                            color=['purple', 'green', 'red'], 
                                                                            drawstyle='steps')
    plt.xlabel('')
    plt.ylabel('persons')
    plt.title(f'COVID-19 in {country}')
```
